## PCA Visualization

It is difficult to visualize high dimensional data. Let's use PCA to find the first two principle components and visualize the data in the two-dimensional space with a single scattered plot.

Scale our data such that feature has single unit of variance

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

PCA with Scikit learn uses very similar approach to other preprocessing functions that come from Scikit learn. We instantiate PCA object, find the principle components using the fit method, then apply the dimensionality reduction by calling transform().

We can also specify how many componenets we can keep when creating the PCA object 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='plasma')
plt.xlabel('First Principle Component')
plt.ylabel('Second Principle Component')
```

Clearly by using these two components we can easily seperate two classes.

## Interpreting the components

Dimensionality reduction comes with the cost of explainability.

The components corresponds to combinations of original features, the components themselves are stored as an attribute of the fitted PCA object

```python
pca.components_
```

In this numpy matrix array, each row represents a principle component and each column relates back to the original features
We can visulize this relationship with heat map

```python
df_comp = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp.head(5)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```

This heatmap and the color bar basically represents the correlation between the various feature and principle component
